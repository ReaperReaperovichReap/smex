```python
import cv2
import os

def extract_frames(video_path, output_dir, frame_interval=30):
    """
    Извлекает кадры из видео и сохраняет их как изображения.

    Args:
        video_path (str): Путь к видеофайлу.
        output_dir (str): Папка, куда будут сохраняться кадры.
        frame_interval (int): Сохранять каждый N-ый кадр.
    """
    # Создаем папку, если её нет
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Открываем видео
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Ошибка: Не удалось открыть видео {video_path}")
        return

    frame_count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Сохраняем каждый frame_interval-ый кадр
        if frame_count % frame_interval == 0:
            # Имя файла: frame_0001.jpg, frame_0002.jpg и т.д.
            frame_filename = os.path.join(output_dir, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            saved_count += 1
            if saved_count % 100 == 0:
                print(f"Сохранено кадров: {saved_count}")

        frame_count += 1

    cap.release()
    print(f"Готово! Извлечено {saved_count} кадров из {video_path}")

# --- Пример использования ---
video_file = "1.mp4" # Путь к вашему видео
output_folder = "raw_frames"
extract_frames(video_file, output_folder, frame_interval=30)
```
